```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
adv_expl = np.loadtxt("build/advection_explicit_tbb.csv", skiprows=10,
                 delimiter=",", dtype=str)
```

```python
adv_impl = np.loadtxt("build/advection_implicit_tbb.csv", skiprows=10,
                 delimiter=",", dtype=str)
```

```python
class data:
    def __init__(self, s, n, t):
        self.size=s
        self.N=n
        self.time=t
```

```python
aelist = []
for a in adv_expl:
    size = int(a[0].split('/')[1])
    N = int(a[0].split('/')[2])
    time = float(a[3])
    d = data(size, N, time)
    aelist.append(d)
```

```python
ailist = []
for a in adv_impl:
    size = int(a[0].split('/')[1])
    N = int(a[0].split('/')[2])
    time = float(a[3])
    d = data(size, N, time)
    ailist.append(d)
```

```python
S = list(set([a.size for a in aelist]))
S.sort()
for s in S:
    x = [(d.N,d.time) for d in aelist if d.size == s]
    t1 = x[0][1]
    y = [d[1]/t1 for d in x]
    x = [d[0] for d in x]
    plt.plot([0,1,10], [0,1,10])
    plt.plot(x,y)
plt.grid()
```

```python
S = list(set([a.size for a in ailist]))
S.sort()
for s in S:
    x = [(d.N,d.time) for d in aelist if d.size == s]
    t1 = x[0][1]
    y = [d[1]/t1 for d in x]
    x = [d[0] for d in x]
    plt.plot([0,1,10], [0,1,10])
    plt.plot(x,y)
plt.grid()
```
