```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
```

```python
nx = 101
nt = 1000
c = 1.0
dx = 1.0 / ( nx - 1 )
dt = 1.0 / nt
c1 = c * dt / dx
t = 0
x = np.linspace ( 0.0, 1.0, nx )
u = np.zeros ( nx )
unew = np.zeros ( nx )

def smooth():
    for i in range ( 0, nx ):
        if ( 0.4 <= x[i] and x[i] <= 0.6 ):
            u[i] = 0.5*(-np.cos(10*np.pi*x[i])+1)

def init():
    smooth()

def euler(j):
    t=j*dt
    for i in range ( 0, nx ):
        unew[i] = u[i] - c1 * ( 3.0/2.0*u[i] - 2*u[i-1] +1.0/2.0*u[i-2])

    for i in range ( 0, nx ):
        u[i] = unew[i]
```

```python
fig = plt.figure()
plt.axes(xlim=(0, 1), ylim=(-2, 2))

smooth()
plt.plot(x, u, lw=2)
```
